```python
def get_chall_list(my_api):

    chall = 'https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=' + my_api
    while(True):
        req = requests.get(chall)
        if req.status_code == 200: break
        elif req.status_code == 429: time.sleep(1)
    challenger_df = json_normalize(req.json()['entries'])
    challenger_df = challenger_df.sort_values('leaguePoints',ascending = False)
    return challenger_df[['summonerId', 'summonerName', 'leaguePoints', 'rank', 'wins', 'losses']]
    

def get_grmaster_list(my_api):
    grmaster = 'https://kr.api.riotgames.com/lol/league/v4/grandmasterleagues/by-queue/RANKED_SOLO_5x5?api_key=' + my_api
    while(True):
        req = requests.get(grmaster)
        if req.status_code == 200: break
        elif req.status_code == 429: time.sleep(1)
    grmaster_df = json_normalize(req.json()['entries'])
    grmaster_df = grmaster_df.sort_values('leaguePoints',ascending = False)
    return grmaster_df[['summonerId', 'summonerName', 'leaguePoints', 'rank', 'wins', 'losses']]

def get_master_list(my_api):
    master = 'https://kr.api.riotgames.com/lol/league/v4/masterleagues/by-queue/RANKED_SOLO_5x5?api_key=' + my_api
    while(True):
        req = requests.get(master)
        if req.status_code == 200: break
        elif req.status_code == 429: time.sleep(1)
    master_df = json_normalize(req.json()['entries'])
    master_df = master_df.sort_values('leaguePoints',ascending = False)
    return master_df[['summonerId', 'summonerName', 'leaguePoints', 'rank', 'wins', 'losses']]

def get_dia_list(my_api):
    dia_df = pd.DataFrame()
    div=['I','II','III','IV']
    for d in div:
        page=1
        while(True):
            dia = 'https://kr.api.riotgames.com/lol/league/v4/entries/RANKED_SOLO_5x5/DIAMOND/'+ d +'?page='+ str(page) +'&api_key=' + my_api
            while(True):
                req = requests.get(dia)
                if req.status_code == 200: break
                elif req.status_code == 429: time.sleep(1)
            tmp = json_normalize(req.json())
            if len(tmp)!=205:
                break
            tmp = tmp.sort_values('leaguePoints',ascending = False)
            dia_df = dia_df.append(tmp)
            page+=1
    return dia_df[['summonerId', 'summonerName', 'leaguePoints', 'rank', 'wins', 'losses']]

def get_accountId(summonerId, my_api):
    account_df = pd.DataFrame()
    for Id in tqdm(summonerId):
        time.sleep(0.001)
        accountId = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/'+Id +'?api_key='+my_api
        while(True):
            req = requests.get(accountId)
            if req.status_code == 200: break
            elif req.status_code == 429: time.sleep(3)
            else: print(req.status_code)
        tmp = json_normalize(req.json())
        account_df = account_df.append(tmp)
    account_df.reset_index(drop=True, inplace=True)
    return account_df

def get_matchId(accountId, my_api):
    matchId_df = pd.DataFrame()
    for Id in tqdm(accountId):
        matchId = 'https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/' + Id + '?endIndex=5&api_key=' + my_api
        while(True):
            req = requests.get(matchId)
            if req.status_code == 200: break
            elif req.status_code == 429: time.sleep(3)
        tmp = json_normalize(req.json()['matches'])
        matchId_df = matchId_df.append(tmp)
    matchId_df.drop_duplicates(subset='gameId', keep='last', inplace=True)    
    matchId_df.reset_index(drop=True, inplace=True)
    return matchId_df

def get_avg_df(stats_df):
    avg_list = []
    for col in stats_df.columns:
        avg=sum(stats_df.iloc[:4][col]) / 5
        avg_list.append(avg)
    blue_avg_df = pd.DataFrame([avg_list], columns=stats_df.columns)
    blue_avg_df.add_prefix("avg_")
    
    avg_list = []
    for col in stats_df.columns:
        avg=sum(stats_df.iloc[5:][col]) / 5
        avg_list.append(avg)
    red_avg_df = pd.DataFrame([avg_list], columns=stats_df.columns)
    red_avg_df.add_prefix("avg_")
    
    avg_df = pd.concat([blue_avg_df, red_avg_df])
    avg_df.reset_index(drop=True, inplace=True)
    return avg_df

def get_match_data(matchId, my_api):
    data = pd.DataFrame()
    for Id in tqdm(matchId):
        match = 'https://kr.api.riotgames.com/lol/match/v4/matches/' + str(Id) + '?api_key=' + my_api
        while(True):
            req = requests.get(match)
            if req.status_code == 200: break
            elif req.status_code == 429: time.sleep(3)
        match_df = json_normalize(req.json())
        
        teams_df = pd.DataFrame.from_dict(match_df['teams'][0])
        teams_df.drop(columns=['bans', 'dominionVictoryScore'], inplace=True)
        
        participants_df = pd.DataFrame.from_dict(match_df['participants'][0])
        participants_df.drop(columns=['participantId'], inplace=True)

        
        stats_df=pd.DataFrame()
        for i in range(len(participants_df['stats'])):
            stats_df = stats_df.append(pd.DataFrame(participants_df['stats'][i], index=[0]))
        stats_df = stats_df[['kills', 'deaths', 'assists', 'totalDamageDealt', 'totalDamageDealtToChampions',
                               'totalHeal', 'visionScore', 'timeCCingOthers', 'totalTimeCrowdControlDealt', 
                               'totalDamageTaken', 'goldEarned','neutralMinionsKilled', 'totalMinionsKilled', 'champLevel']]
        
        avg_df = get_avg_df(stats_df)
        test_data = pd.concat([teams_df,avg_df], axis=1)
        blue_data = (test_data.iloc[0]).add_prefix("blue_")
        red_data = (test_data.iloc[1]).add_prefix("red_")
        tmp = pd.concat([blue_data, red_data])
        data = data.append(tmp, ignore_index=True)
    return data

```
